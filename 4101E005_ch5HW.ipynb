```python
#2.1
def largest(*numbers):
  return max(numbers)

print(largest(10, 50, 100, -10, -5))
```

```python
#2.3
def salary(hourlypay, hours, bonus = 0):
  return hourlypay * hours + bonus

print(salary(60, 8))
print(salary(60, 8, 10))
```

```python
#2.5
def sum(P, r, t):
  return P * (1 + r * t)

P = eval(input("Input 本金$"))
r = eval(input("Input 年利率%(float)"))
t = eval(input("Input 年數"))
print("單利本利和 = $", sum(P, r, t))
```

```python
#2.6
def sum(P, r, t):
  return P * pow((1 + r), t)

P = eval(input("Input 本金$"))
r = eval(input("Input 年利率%(float)"))
t = eval(input("Input 年數"))
print("複利本利和 = $", sum(P, r, t))
```
